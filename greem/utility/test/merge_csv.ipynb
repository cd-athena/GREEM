```python
import pandas as pd
import polars as pl
import os
```

```python
input_csvs: list[str] = [d for d in os.listdir() if d.endswith('.csv') and not 'test' in d]
input_csvs
```

```python
monitoring_stream_df = pd.read_csv(input_csvs[0], index_col=0)
monitoring_stream_df.head(3)
```

```python
idle_df = pd.read_csv(input_csvs[2], index_col=0)
idle_df
```

```python

idle_energy_per_second = float(idle_df.transpose()['cpu_energy_per_second'][0])
idle_energy_per_second
```

```python
encoding_results = pd.read_csv(input_csvs[1], index_col=0)
encoding_results.head(3)
```

```python
monitoring_stream_df.head(2)
```

```python
idle_df_seconds = float(idle_df.transpose()['cpu_energy']) / float(idle_df.transpose()['cpu_energy_per_second'])
```

```python
gpu_idle_energy_per_second = float(idle_df.transpose()['gpu_energy'][0]) / idle_df_seconds
mem_idle_energy_per_second = float(idle_df.transpose()['ram_energy'][0]) / idle_df_seconds
```

```python
encoding_results['idle_energy.duration.cpu'] = encoding_results['duration'] * idle_energy_per_second
encoding_results['idle_energy.duration.gpu'] = encoding_results['duration'] * gpu_idle_energy_per_second
encoding_results['idle_energy.duration.mem'] = encoding_results['duration'] * mem_idle_energy_per_second
```

```python

df_colums: list[str] = ['fan_speed', 'fb_memory_usage.used', 'fb_memory_usage.free',
        'utilization.gpu_util', 'utilization.memory_util',
        'temperature.gpu_temp',
        'clocks.graphics_clock', 'clocks.sm_clock', 'clocks.mem_clock',
        ]

df_value_dict: dict[
    int, dict[str, object]
    ] = dict()

for index, row in encoding_results.iterrows():
    
    df_value_dict[index] = dict()
    
    video_name = row['video_name']
    bitrate = row['bitrate']
    
    temp_df = monitoring_stream_df.loc[
        (monitoring_stream_df['current_video'] == video_name) &
        (monitoring_stream_df['bitrate'] == bitrate)
        ]
    
    monitored_value_count: int = len(temp_df)
    
    df_value_dict[index].update({'value.count': monitored_value_count})
    df_value_dict[index].update({'value.idle': monitored_value_count})
    
    describe_df = temp_df[df_colums].describe()
    for col in df_colums:
        
        mean = describe_df[col].mean()
        min = describe_df[col].min()
        max = describe_df[col].max()
        std = describe_df[col].std()
        
        temp_dict = {
            f'{col}.mean': mean, f'{col}.min': min, f'{col}.max': max
        }
        
        df_value_dict[index].update(temp_dict)
```

```python
copy_df = pd.DataFrame.from_dict(df_value_dict).transpose()
copy_df = copy_df.convert_dtypes({'value.count': int})
copy_df.head(3)
```

```python
merge_df = pd.merge(encoding_results, copy_df, left_index=True, right_index=True)
merge_df
```

```python
merge_df.to_csv('test.csv')
```
